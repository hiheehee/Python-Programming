## pivot
* dataframe의 형태를 변경한다.
* 인덱스, 컬럼, 데이터로 사용할 컬럼을 명시한다.

* 데이터에 동일 date에 A값이 2개 이상 존재한다면 아래 오류가 발생한다.
    - 'ValueError: Index contains duplicate entries, cannot reshape'

```python
df.pivot(index='지역', columns = '요일')
```

```python
pd.pivot_table(df, index='요일', columns='지역', aggfunc=np.mean)
```
