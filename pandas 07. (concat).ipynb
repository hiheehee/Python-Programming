```python
import pandas as pd
import numpy as np
```

## dataframe 병합하기
* pandas.concat 함수를 사용해 dataframe을 병합할 수 있다.
* 축을 따라 dataframe을 병합 가능하다.
    - 기본 axis = 0 -> 행단위 병합

```python
df1 = pd.DataFrame({'key1' : np.arange(10), 'value1' : np.random.randn(10)})
df1
```

```python
df2 = pd.DataFrame({'key1' : np.arange(10), 'value1' : np.random.randn(10)})
df2
```

```python
pd.concat([df1, df2], ignore_index=True)
```

```python
pd.concat([df1, df2], ignore_index=False)
```

```python
pd.concat([df1, df2], axis=1)
```

```python
pd.concat([df1, df2], axis=0)
```

```python
df3 = pd.DataFrame({'key2' : np.arange(10), 'value2' : np.random.randn(10)})
df3
```

```python
pd.concat([df1, df3])
```
