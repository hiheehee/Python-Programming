```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## 그래프 데이터 생성

```python
x = np.linspace(0, 10, 11)
y = x ** 2 + x + 2 + np.random.randn(11)

print(x)
print(y)
```

## 그래프 출력하기
* plot함수 (선 그래프), scatter(점 그래프), hist(히스토그램) 등 사용한다.
* 함수의 parameter 혹은 plt의 다른 함수로 그래프 형태 및 설정을 변경 가능하다.
* 기본적으로, x, y에 해당하는 값이 필요하다.

```python
plt.plot(x, y)
```

```python
plt.scatter(x, y)
```

## 그래프에 주석 추가
* 타이틀 title
* x축 이름 xlabel
* y축 이름 ylabel
* 모눈선 grid(True)
* x축 범위 지정 xlim(n,m)
* y축 범위 지정 ylim(n,m)

```python
plt.title('X-Y relation')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.grid(True)
plt.xlim(0, 20)
plt.ylim(0, 200)
plt.plot(x, y)
```

## plot 함수 parameters
* 그래프의 형태에 대한 제어가 가능하다.
* [plot함수 도큐먼트](https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot)
* 그래프의 색상 변경
* 그래프 선스타일 변경
* 그래프 두께 변경
* [plot 함수 keyword parameter](https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot)

```python
plt.plot(x, y, color='black', 
          linestyle='--', marker='^',
         markerfacecolor='red', markersize=10)
```

## subplot
* 구획을 나눠 각각의 subplot에 그래프를 출력할 수 있다.
* plt.subplot(행의 수, 열의 수, 인덱스)

```python
plt.subplot(3, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(3, 2, 2)
plt.plot(x, y*2, 'g')

plt.subplot(3, 2, 3)
plt.plot(y, x, 'k')

plt.subplot(3, 2, 4)
plt.plot(x, np.exp(x), 'b')

plt.subplot(3, 2, 5)
plt.plot(y*2, x, 'y')

plt.subplot(3, 2, 6)
plt.plot(np.exp(x), x, 'pink')
```

## hist
* histogram을 생성할 수 있다.
* bins로 historgram bar 개수를 설정한다.

```python
data = np.random.randint(1, 100, size=200)
print(data)
```

```python
plt.hist(data, bins=20, alpha=0.3)
plt.xlabel('값')
plt.ylabel('개수')
plt.grid(True)
```
