# DataFrame
* Series가 1차원이라면 DataFrame은 2차원으로 확대된 버젼이다.
* Excel spreadsheet이라고 생각하면 이해하기 쉽다.
* 2차원이기 때문에 인덱스가 row, column로 구성된다.
* row는 각 개별 데이터를, column은 개별 속성을 의미한다.
* Data Analysis, Machine Learning에서 data 변형을 위해 가장 많이 사용한다.

## head
* 데이터의 처음부터 n개만 보기 위한 함수를 의미한다.

## tail
* 데이터의 끝에서 n개반 보기위한 함수를 의미한다.

## dataframe 데이터 파악하기
* shape 속성 (row, column)
* describe 함수 - 숫자형 데이터의 통계치 계산
* info 함수 - 데이터 타입, 각 아이템의 개수 등 출력

###  데이터 타입
* 숫자형(Numerical Type) 데이터
    - 연속성을 띄는 숫자로 이루어진 데이터
        - 예) Age, Fare 등
* 범주형(Categorical Type) 데이터
    - 연속적이지 않은 값(대부분의 경우 숫자를 제외한 나머지 값)을 갖는 데이터
        - 예) Name, Sex, Ticket, Cabin, Embarked
    - 어떤 경우, 숫자형 타입이라 할지라도 개념적으로 범주형으로 처리해야할 경우가 있음
        - 예) Pclass
* astype함수 : 데이터 타입 변경

## NaN 처리 방법
* dropna 함수 : 데이터에서 삭제할 수 있다.
* fillna 함수 : 다른 데이터 값으로 치환할 수 있다.
    - 평균으로 대체하기

```python
train_data.dropna()
```

```python
# Age', 'Cabin'열값 중 NaN값 삭제
train_data.dropna(subset=['Age', 'Cabin'])
```

```python
# NaN값을 가진 'Age', 'Cabin'열 삭제
train_data.dropna(axis=1)
```

```python
# 생존자 나이 평균
mean1 = train_data[train_data['Survived'] == 1]['Age'].mean()

# 사망자 나이 평균
mean0 = train_data[train_data['Survived'] == 0]['Age'].mean()

print(mean1, mean0)
```

```python
train_data[train_data['Survived'] == 1]['Age'].fillna(mean1)
train_data[train_data['Survived'] == 0]['Age'].fillna(mean0)
```

```python
train_data.loc[train_data['Survived'] == 1, 'Age'] = train_data[train_data['Survived'] == 1]['Age'].fillna(mean1)
train_data.loc[train_data['Survived'] == 0, 'Age'] = train_data[train_data['Survived'] == 0]['Age'].fillna(mean0)
```

## One-hot encoding
* 범주형 데이터는 분석단계에서 계산이 어렵기 때문에 숫자형으로 변경이 필요하다.
* 범주형 데이터의 각 범주(category)를 column레벨로 변경한다.
* 해당 범주에 해당하면 1, 아니면 0으로 채우는 인코딩 기법
    * pandas.get_dummies 함수 사용
        * drop_first가 True면 첫번째 카테고리 값은 사용하지 않고 False면 사용한다.

```python
pd.get_dummies(train_data, columns = ['Sex', 'Embarked'], drop_first=True)
```

```python
pd.get_dummies(train_data, columns = ['Sex', 'Embarked'], drop_first=False)
```
